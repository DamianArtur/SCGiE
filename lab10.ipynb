```python
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(a, x, N=100):
    hist = []
    for _ in range(N):
        hist.append(x)
        x = a * x*(1 - x)
    return hist
```

```python
# epsilon maszynowy dla typu float w pythonie
print(sys.float_info.epsilon)
```

```python
a = 3.7

# wartości początkowe
x0_1 = 0.1234567890123456
x0_2 = 0.1234567890123457

hist1 = np.array(fun(a, x0_1))
hist2 = np.array(fun(a, x0_2))

# poliecznie różnicy otrzymanych trajektorii
diff = np.abs(hist1 - hist2)
```

```python
plt.plot(diff)
plt.title('Wykres różnic dla otrzymanych trajektorii')
plt.ylabel('Różnica')
plt.xlabel('Krok')
plt.grid(True)
plt.show()
```
